# FIND SONG FROM MIC

```python
import asyncio
import csv
from datetime import datetime
from shazamio import Shazam
import pyaudio
from pydub import AudioSegment
import keyboard

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100

details = {}
is_song_found = False

async def find_song():
    global details, is_song_found
    shazam = Shazam()
    p = pyaudio.PyAudio()
    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)

    while not is_song_found:
        if keyboard.is_pressed('q'):  # Check if the 'q' key is pressed
            print('Stopping song recognition...')
            break

        data = stream.read(CHUNK)
        audio = AudioSegment(
            data=data,
            sample_width=p.get_sample_size(FORMAT),
            frame_rate=RATE,
            channels=CHANNELS
        )
        out = await shazam.recognize_song(audio)
        if out.get("track"):
            details = out
            print(f"Song recognized: {details['track']['title']} by {details['track']['subtitle']}")
            save_to_csv(details['track']['title'], details['track']['subtitle'])
            is_song_found = True
            break

def save_to_csv(title, artist):
    now = datetime.now()
    timestamp = now.strftime("%Y-%m-%d %H:%M:%S")
    with open('song_history.csv', mode='a', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow([timestamp, title, artist])

asyncio.run(find_song())
```
